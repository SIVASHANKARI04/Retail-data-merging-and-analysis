```python
import pandas as pd
```

```python
df_stores=pd.read_csv("/content/stores_data_set (1).csv")
df_sales=pd.read_csv("/content/sales_data_set (1).csv")
df_features=pd.read_csv("/content/Features_data_set (1).csv")
```

```python
df_stores.head()
```

```python
df_stores.shape
```

```python
df_sales.head()
```

```python
df_sales.shape
```

```python
df_sales_stores=pd.merge(df_sales,df_stores,on="Store")
```

```python
df_sales_stores.head()
```

```python
df_features.head()
```

```python
df_features.shape
```

```python
df_features.drop(columns="IsHoliday",axis=1,inplace=True)
```

```python
df_features.head()
```

```python
df_final=pd.merge(df_sales_stores,df_features,on=["Store","Date"])
```

```python
df_final.head()
```

```python
df_final.shape
```

```python
df_final.info()
```

```python
df_final["Date"]=pd.to_datetime(df_final["Date"],format="%d/%m/%Y")
```

```python
df_final.info()
```

```python
df_final["year"]=df_final["Date"].dt.year
df_final["month"]=df_final["Date"].dt.month
df_final["day"]=df_final["Date"].dt.day
df_final["day_of_week"]=df_final["Date"].dt.day_of_week
df_final["week"]=df_final["Date"].dt.isocalendar().week
```

```python
df_final
```

# 1.Identify average customer visit in Type C store in April Month.


```python
 df_final[(df_final["Type"]=="C") & (df_final["month"]==4)]["Weekly_Sales"].mean()
```

# 2.Find the highest sales in each store type on Holiday season.


```python
df_final[df_final["IsHoliday"]=="True"].groupby("Type")["Weekly_Sales"].max()
```

# 3.What is the expected sales of each depeartment when the unemployment factor is greater than 7.


```python
df_final[df_final["Unemployment"]>7].groupby("Dept")["Weekly_Sales"].mean()
```

# 4.Aggregate net sales for each department in month wise.


```python
df_final.groupby(["Dept","month"])["Weekly_Sales"].sum()
```

# 5.Identify the performance of the store on weekwise, without offers.


```python
df_final[df_final["Weekly_Sales"]>0].groupby(["Store","week"])["Weekly_Sales"].sum()
```
